## Bar Plots
Here is the [`Documentation`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html) for Bar

```python
np.random.seed(5)
N = 10
x = range(N)
y = np.random.random(N) *100
y_err = np.random.randn(N) *10
y1 = np.random.random(N) *100
y1_err = np.random.randn(N) *10
width = 0.4
bottom = 0.6
ticksLables = ['One','Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten']

#  A Simple two sided Bar Graph
fig, ax = plt.subplots(nrows=2, figsize=(10, 10), sharex=True) # sharex for sharing the same x axis for both the plots
ax[0].bar(x,y, -width, align='edge', alpha=0.5, label='left', yerr=y_err,color='blue')
ax[0].bar(x,y1, width, align='edge', alpha=0.5, color='green',linewidth=2.5, label='right', yerr=y1_err)
ax[0].legend()
# ax[0].set_xticks(x,rotation=30)
# ax[0].set_xticklabels(ticksLables)
plt.xticks(x, ticksLables, rotation=30)
ax[0].set_title('Bar Graph 1')

plt.subplots_adjust(wspace=0, hspace=0) # for sticking those two graphs horizontalliy 

###########################################

# Simple stacked bar graph
ax[1].bar(x,y, color='blue', alpha=0.5, label='left')
ax[1].bar(x,y1, bottom=y, color='green', alpha=0.5, label='right')
```

```python

people = ('A','B','C','D','E','F','G','H')
segments = 4

# generate some multi-dimensional data & arbitrary labels
data = 3 + 10* np.random.rand(segments, len(people))
percentages = (np.random.randint(5,20, (len(people), segments)))
y_pos = np.arange(len(people))

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

colors ='rgbwmc'
patch_handles = []
left = np.zeros(len(people)) # left alignment of data starts at zero
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d, 
      color=colors[i%len(colors)], align='center', 
      left=left))
    # accumulate the left-hand offsets
    left += d

# go through all of the bar segments and annotate
for j in xrange(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (percentages[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Distance')

plt.show()
```
